```python
# prompt: â€¢ Filtering and Sorting:
# 1. Filter the dataset to show only the countries with a Score greater than 7.5.
# 2. For the filtered dataset - Sort the dataset by GDP per Capita in descending order and display the
# top 10 rows.

# Filter for countries with a Score greater than 7.5
print("The dataset to show only the countries with a Score greater than 7.5 are :")
filtered_df = df[df['score'] > 7.5]

# Sort the filtered dataset by GDP per Capita in descending order and display the top 10
# The column name to the correct name from the dataset: "Log GDP per capita"
top_10_gdp = filtered_df.sort_values('Log GDP per capita', ascending=False).head(10)
top_10_gdp
```

```python
# prompt: Adding New Columns:
# 1. Create a new column called Happiness Category that categorizes countries into three categories
# based on their Score:
# Low âˆ’ (Score < 4)
# Medium âˆ’ (4 â‰¤ Score â‰¤ 6)
# High âˆ’ (Score > 6
# get low, medium and high hapiness category of 2 countries each category

# Create the 'Happiness Category' column
def categorize_happiness(score):
    if score < 4:
        return 'Low'
    elif 4 <= score <= 6:
        return 'Medium'
    else:
        return 'High'

df['Happiness Category'] = df['score'].apply(categorize_happiness)

# Get two countries for each happiness category
low_happiness_countries = df[df['Happiness Category'] == 'Low'].head(2)
medium_happiness_countries = df[df['Happiness Category'] == 'Medium'].head(2)
high_happiness_countries = df[df['Happiness Category'] == 'High'].head(2)


print ("A new column called Happiness Category that categorizes countries into three categories based on their score are:\n")
# Display the results
print("Low Happiness Countries:\n")
print(low_happiness_countries[['Country name', 'score']])
print("\nMedium Happiness Countries:\n")
print(medium_happiness_countries[['Country name', 'score']])
print("\nHigh Happiness Countries:\n")
print(high_happiness_countries[['Country name', 'score']])
```

```python
# prompt: 3.2 Problem - 2 - Some Advance Data Exploration Task:
# Task - 1 - Setup Task - Preparing the South-Asia Dataset:
# Steps:
# 1. Define the countries in South Asia with a list for example:
# south asian countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
# "Maldives", "Nepal", "Pakistan", "Srilanka"]
# 2. Use the list from step - 1 to filtered the dataset {i.e. filtered out matching dataset from list.}
# 3. Save the filtered dataframe as separate CSV files for future use.

# Import pandas
import pandas as pd

# Define the list of South Asian countries
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                        "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Load the DataFrame (Make sure 'WHR-2024-5CS037.csv' is in the correct path)
df = pd.read_csv("WHR-2024-5CS037.csv")  # This line is crucial

# Filter the DataFrame to include only South Asian countries
south_asia_df = df[df['Country name'].isin(south_asian_countries)]

# Save the filtered DataFrame to a CSV file
south_asia_df.to_csv('south_asia_happiness.csv', index=False)
```

```python
# prompt: Task - 2 - Composite Score Ranking:
# Tasks:
# 1. Using the SouthAsia DataFrame, create a new column called Composite Score that combines the
# following metrics:
# Composite Score = 0.40 Ã— GDP per Capita + 0.30 Ã— Social Support
# + 0.30 Ã— Healthy Life Expectancy
# 2. Rank the South Asian countries based on the Composite Score in descending order.
# 3. Visualize the top 5 countries using a horizontal bar chart showing the Composite Score.
# 4. Discuss whether the rankings based on the Composite Score align with the original Score - support your
# discussion with some visualization plot.

# Calculate the Composite Score
south_asia_df['Composite Score'] = (0.40 * south_asia_df['Log GDP per capita'] +
                                   0.30 * south_asia_df['Social support'] +
                                   0.30 * south_asia_df['Healthy life expectancy'])

# Rank the countries based on the Composite Score
south_asia_df = south_asia_df.sort_values('Composite Score', ascending=False)

# Visualize the top 5 countries using a horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Composite Score', y='Country name', data=south_asia_df.head(5), palette='viridis')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.show()

# Discussion and visualization comparing Composite Score and original Score
plt.figure(figsize=(10, 6))
plt.scatter(south_asia_df['score'], south_asia_df['Composite Score'])
for i, row in south_asia_df.iterrows():
    plt.annotate(row['Country name'], (row['score'], row['Composite Score']))

plt.xlabel('Original Happiness Score')
plt.ylabel('Composite Score')
plt.title('Comparison of Original Happiness Score and Composite Score')
plt.show()

# Discussion (Example):
# "The scatter plot shows the relationship between the original happiness score and the composite score.
#  We can observe that there's [a positive/negative/no clear] correlation between the two.
#  Some countries rank higher/lower in the composite score compared to the original score.
#  For example [country A] has a high composite score but a relatively low happiness score because [mention contributing factor from composite score calculation e.g., high GDP per capita],
#  while [country B] has a lower composite score despite having a higher happiness score because [mention contributing factor e.g., low GDP per capita].
#  These differences highlight how the composite score prioritizes specific aspects of wellbeing like economic factors and health, while the original happiness score might include other elements not captured by our composite score."
```

```python
# prompt: Task - 3 - Outlier Detection:
# Tasks:
# 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
# 2. Define outliers using the 1.5 Ã— IQR rule.
# 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
# in a different color.
# 4. Discuss the characteristics of these outliers and their potential impact on regional averages.

# Calculate IQR for 'Log GDP per capita' and 'score'
Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Define outlier thresholds
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identify outliers
outliers = south_asia_df[
    (south_asia_df['Log GDP per capita'] < lower_bound_gdp) |
    (south_asia_df['Log GDP per capita'] > upper_bound_gdp) |
    (south_asia_df['score'] < lower_bound_score) |
    (south_asia_df['score'] > upper_bound_score)
]

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], label='Non-Outliers')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers')

# Annotate outlier points with country names
for i, row in outliers.iterrows():
  plt.annotate(row['Country name'], (row['Log GDP per capita'], row['score']))

plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Outlier Detection in South Asian Countries')
plt.legend()
plt.show()

# Discuss the characteristics and potential impact of the outliers (Example)
print("Outlier Analysis:")
if not outliers.empty:
    print(outliers[['Country name', 'Log GDP per capita', 'score']])
    print("\nDiscussion:")
    print("The identified outliers deviate significantly from the other South Asian countries in terms of their GDP per capita and/or happiness score.")
    print("These outliers can significantly skew regional averages, making it crucial to consider their unique characteristics when analyzing South Asian trends.")
    # Add more specific discussion based on the outlier countries
else:
    print("No outliers were identified based on the defined criteria.")
```

```python
# prompt: Task - 4 - Exploring Trends Across Metrics:
# Tasks:
# 1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
# {pearson correlation} with the Score for South Asian countries.
# 2. Create scatter plots with trendlines for these metrics against the Score.
# 3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
# South Asian countries.

import matplotlib.pyplot as plt
import seaborn as sns

# Filter for South Asian countries
south_asia_df = df[df['Country name'].isin(south_asian_countries)]

# Choose two metrics (replace with your desired metrics)
metric1 = 'Freedom to make life choices'
metric2 = 'Generosity'

# Calculate Pearson correlation
correlation1 = south_asia_df[metric1].corr(south_asia_df['score'], method='pearson')
correlation2 = south_asia_df[metric2].corr(south_asia_df['score'], method='pearson')

print(f"Pearson correlation between {metric1} and Score: {correlation1}")
print(f"Pearson correlation between {metric2} and Score: {correlation2}")

# Create scatter plots with trendlines
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.regplot(x=metric1, y='score', data=south_asia_df)
plt.title(f'{metric1} vs. Score')

plt.subplot(1, 2, 2)
sns.regplot(x=metric2, y='score', data=south_asia_df)
plt.title(f'{metric2} vs. Score')

plt.tight_layout()
plt.show()

# Discuss strongest and weakest relationships
print("\nDiscussion:")
if abs(correlation1) > abs(correlation2):
    print(f"The strongest relationship is between {metric1} and Score, with a correlation of {correlation1}.")
    print(f"The weakest relationship is between {metric2} and Score, with a correlation of {correlation2}.")
else:
    print(f"The strongest relationship is between {metric2} and Score, with a correlation of {correlation2}.")
    print(f"The weakest relationship is between {metric1} and Score, with a correlation of {correlation1}.")
# Add more detailed discussion based on the observed correlations and scatter plots.
```

```python
# prompt: Task - 5 - Gap Analysis:
# Tasks:
# 1. Add a new column, GDP-Score Gap, which is the difference between GDP per Capita and the Score
# for each South Asian country.
# 2. Rank the South Asian countries by this gap in both ascending and descending order.
# 3. Highlight the top 3 countries with the largest positive and negative gaps using a bar chart.
# 4. Analyze the reasons behind these gaps and their implications for South Asian countries.

# Calculate the GDP-Score Gap
south_asia_df['GDP-Score Gap'] = south_asia_df['Log GDP per capita'] - south_asia_df['score']

# Rank the countries by the gap
south_asia_df_ascending = south_asia_df.sort_values('GDP-Score Gap')
south_asia_df_descending = south_asia_df.sort_values('GDP-Score Gap', ascending=False)

# Display the rankings (optional)
print("Ascending order by GDP-Score Gap:\n", south_asia_df_ascending[['Country name', 'GDP-Score Gap']])
print("\nDescending order by GDP-Score Gap:\n", south_asia_df_descending[['Country name', 'GDP-Score Gap']])

# Highlight top 3 countries with largest positive and negative gaps
top3_positive_gap = south_asia_df_descending.head(3)
top3_negative_gap = south_asia_df_ascending.head(3)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(top3_positive_gap['Country name'], top3_positive_gap['GDP-Score Gap'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('GDP-Score Gap')
plt.title('Top 3 Countries with Largest Positive Gaps')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
plt.bar(top3_negative_gap['Country name'], top3_negative_gap['GDP-Score Gap'], color='lightcoral')
plt.xlabel('Country')
plt.ylabel('GDP-Score Gap')
plt.title('Top 3 Countries with Largest Negative Gaps')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

# Analyze the reasons (example - replace with your own analysis)
print("Analysis of GDP-Score Gaps:")
print("Large positive gaps might indicate that countries have high GDP but lower happiness scores possibly due to income inequality, corruption, or lack of social safety nets.")
print("Large negative gaps may indicate that countries with lower GDP per capita have other factors contributing to higher happiness, like strong social support, community bonds, or a focus on wellbeing over economic growth.")
```
